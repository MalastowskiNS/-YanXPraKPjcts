```python
num_col_ferma = ferma.select_dtypes(include=np.number).columns.values
n = len(num_col_ferma)
fig, ax = plt.subplots(1, n, figsize=(18, 6))
fig.suptitle('"Ящик с усами" для количественных параметров коров ассоциации пастбищ «ЭкоФерма»')
boxprops = dict(facecolor = "lightblue")
for i in range(n):
    ferma.boxplot(num_col_ferma[i], ax=ax[i], patch_artist = True, boxprops = boxprops)
plt.subplots_adjust(wspace=0.5) 
plt.show()
```

#### 4. Обучение модели линейной регрессии

**1. Постройте модель линейной регрессии на данных датасета**

Создадим функцию для построение линейной модели регрессии на базе PipeLine:

```python
def lin_reg(data, X_num_ferma, X_cat_ferma, target):
    y = data[target]
    X = pd.concat([data[X_num_ferma], data[X_cat_ferma]], axis=1)
    RANDOM_STATE = 42
    X_train, X_test, y_train, y_test = train_test_split(
        X, 
        y, 
        random_state=RANDOM_STATE)
    # обработка численных при
    num_features = X_num_ferma.tolist()
    # заполнение пропусков, масштабирование
    num_steps = [('scaler', StandardScaler())]
    num_transformer = Pipeline(num_steps)
    # обработка категориальных признаков
    cat_features = X_cat_ferma.tolist()
    # преобразование категориальных признаков
    cat_steps = [('encoder', OneHotEncoder(handle_unknown = 'ignore', drop='first'))]
    cat_transformer = Pipeline(cat_steps)
    # собираем все вместе
    preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features)])
    steps = [('preprocessor', preprocessor), ('regressor', LinearRegression(n_jobs=-1))]
    RegMod= Pipeline(steps)
    RegMod.fit(X_train, y_train)
    return X_train, X_test, y_train, y_test, RegMod
```

Для первой модели на основании проведенного анализа используем следущие характеристики:
1. Количественные признаки:
   - Параметры кормления **ЭКЕ**(0.57 коэф. корр), **Сырой протеин**(0.3 коэф. корр), **СПО**(0.6 коэф. корр);
   - Характеристики молока: жирность(0.49 коэф. корр), белок(~0.01 коэф. корр) - можно не использовать хар-ки молока.
2. Качественые признаки:
   - Особенности коровы: **Порода**, **Возраст**, **Имя Папы**, **Тип пастбища**, **порода папы** - показано, что возраст и порода быка значительно влияют на корреляции количественных параметров, отстальные проверим в процессе сборки модели;
3. Целевая функция - **удой**

```python
X_num_ferma = num_col_ferma[~np.isin(num_col_ferma, ['milk_yield', '!energy_feed_unit', '!protein_abs', '!sugar_protein_ratio', 'fat', 'protein'])]
X_cat_ferma = cat_col_ferma[~np.isin(cat_col_ferma, ['species', 'pasture_type', '!bull_species', 'taste', '!age','bull_name'])]
target = 'milk_yield'
X_train, X_test, y_train, y_test, frst_model = lin_reg(ferma, X_num_ferma, X_cat_ferma, target)
```

```python
def outliers_drop (data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_lim = Q3 + 1.5*IQR
    lower_lim = Q1 - 1.5*IQR
    data=data.loc[(data[col]<=upper_lim)&(data[col]>=lower_lim)]
    return data
```

#### 5. Обучение модели логистической регрессии

**1. Обучить модель логистической регресс LogisticRegression**

```python
def lin_log(data, X_num_ferma, X_cat_ferma, target):
    y = data[target]
    X = pd.concat([data[X_num_ferma], data[X_cat_ferma]], axis=1)
    RANDOM_STATE = 42
    X_train, X_test, y_train, y_test = train_test_split(
        X, 
        y, 
        random_state=RANDOM_STATE, stratify=y)
    # обработка численных при
    num_features = X_num_ferma.tolist()
    # заполнение пропусков, масштабирование
    num_steps = [('scaler', StandardScaler())]
    num_transformer = Pipeline(num_steps)
    # обработка категориальных признаков
    cat_features = X_cat_ferma.tolist()
    # преобразование категориальных признаков
    cat_steps = [('encoder', OneHotEncoder(handle_unknown = 'ignore',  drop='first'))]
    cat_transformer = Pipeline(cat_steps)
    # собираем все вместе
    preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features)])
    step_prep = ('preprocessor', preprocessor)
    step_sol  = ('regressor', LogisticRegression(n_jobs=-1))
    #step_sol  = ('regressor', RandomForestRegressor())
    LogMod= Pipeline([step_prep, step_sol])
    LogMod.fit(X_train, y_train)
    return X_train, X_test, y_train, y_test, LogMod
```

Для модели на основании модели регрессии используем следущие характеристики:
1. Количественные признаки:
   - Параметры кормления **ЭКЕ**, **Сырой протеин**, **ЭКЕ^2**;
   - Характеристики молока: **жирность**, **белок**
2. Качественые признаки:
   - Особенности коровы: **Порода**, **Возраст**, **Имя Папы**, **Тип пастбища**, **Порода папы**, **СПО_кат**, **Вид** ;
3. Целевая функция - **вкус молока**

Можно построить матрицу ошибок и оценить качество модели на тестовой выборке по метрикам accuracy, recall, precision

```python
def conf_matrix_plot(y_true, probability, trhld):
    y_predict = (probability>=trhld).astype(int)
    cm = confusion_matrix(y_true, y_predict)
    plt.figure(figsize=(10, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
    plt.ylabel('True label')
    plt.xlabel('Predicted');
    plt.suptitle('Матрица ошибок модели Логистической регрессии');
    plt.show()
    recall = recall_score(y_true, y_predict)
    precision = precision_score(y_true, y_predict)
    print('Метрика accuracy: ',round(accuracy_score(y_predict, y_true) ,3))
    print('Метрика recall: ',round(recall ,3))
    print('Метрика precision: ',round(precision,3))
```

```python
def outliers_cat (data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_lim = Q3 + 1.5*IQR
    lower_lim = Q1 - 1.5*IQR
    data = data[col].apply(lambda x:  "1" if x < lower_lim else ( "3" if x > upper_lim else "2" ))
    return data
```
