# Исследование Интернет-магазина «В один клик»

**Задача: Персонализировать предложения постоянным клиентам Интернет-магазина «В один клик», чтобы увеличить их покупательскую активность**

**Цели:**

**Разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность** 

    


**Заказчиком предоставлены следующие данные в таблицах:**
*  **market_file**:

       - id — номер покупателя в корпоративной базе данных;
       - Покупательская активность  — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень»;
       - Тип сервиса  — уровень сервиса, например «премиум» и «стандарт»;
       - Разрешить сообщать  — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель;
       - Маркет_актив_6_мес  — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента;
       - Маркет_актив_тек_мес  — количество маркетинговых коммуникаций в текущем месяце;
       - Длительность — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте;
       - Акционные_покупки  — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев;
       - Популярная_категория — самая популярная категория товаров у покупателя за последние 6 месяцев;
       - Средний_просмотр_категорий_за_визит  — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца;
       - Неоплаченные_продукты_штук_квартал  — общее число неоплаченных товаров в корзине за последние 3 месяца;
       - Страниц_за_визит  — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца;
       - Ошибка_сервиса  — число сбоев, которые коснулись покупателя во время посещения сайта.
   
*  **market_money**:
  
       - id — номер покупателя в корпоративной базе данных;
       - Период — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц';
       - Выручка — сумма выручки за период.         
      
*  **market_time**:
  
       - id— номер покупателя в корпоративной базе данных;
       - Период  — название периода, во время которого зафиксировано общее время;
       - минут — значение времени, проведённого на сайте, в минутах.

*  **money**:
  
       - id— номер покупателя в корпоративной базе данных;
       - Прибыль — значение прибыли.

```python
def boxplot_tab(data, cols, title):
    n = len(cols)
    fig, ax = plt.subplots(1, n, figsize=(25, 6))
    fig.suptitle(title)
    boxprops = dict(facecolor = "lightblue")
    for i in range(n):
        data.boxplot(cols[i], ax=ax[i], patch_artist = True, boxprops = boxprops)
        plt.subplots_adjust(wspace=0.5)
    plt.show()
```

Масштабирование и кодирование параметров:

У нас есть следующие типы параметров:
 * Категориальный номинальный: 'Тип сервиса', 'Разрешить сообщать', 'Популярная_категория' (используем OneHotEncoder, OrdinalEncoder);
 * Категориальный порядковый: нет;
 * Количественный непрерывный:'Маркет_актив_6_мес', 'Длительность',  'Страниц_за_визит', 'препредыдущий_месяц_money'   'предыдущий_месяц_money', 'текущий_месяц_money    'предыдущий_месяц_time', 'текущий_месяц_time' (используем Standard Scaler, MinMax Scaler);
 * Количественный дискретный: 'Акционные_покупки', "Маркет_актив_тек_мес', "Средний_просмотр_категорий_за_визит", "Неоплаченные_продукты_штук_квартал", "Ошибка_сервиса" (используем Standard Scaler, MinMax Scaler).

Создадим pipeline

```python
def prediction_model(data, X_num, X_cat, target, NumScaler, CatCoder, model_step):
    RANDOM_STATE = 42
    TEST_SIZE = 0.25
    
    # выбор масшабирования
    if NumScaler == 'Standart':
        Scaler = StandardScaler()
    else:
        Scaler = MinMaxScaler()
    # выбор кодирования
    if CatCoder == 'OneHot':
        Coder = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
    else:
        Coder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    y = data[target]
    X = pd.concat([data[X_num], data[X_cat]], axis=1)
    # разбивка входных данных
    X_train, X_test, y_train, y_test = train_test_split(
        X, 
        y, 
        test_size = TEST_SIZE,
        random_state=RANDOM_STATE, stratify=y)
    # обработка входных данных
    num_features = X_num.tolist()
    cat_features = X_cat.tolist()
    # преобразование категориальных признаков
    cat_steps = [('SimpleImputer_cat', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                 ('Coder', Coder)]
    # преобразование численных признаков
    num_steps = [('SimpleImputer_num', SimpleImputer(missing_values=np.nan, strategy='median')),
                 ('Scaler', Scaler)]
    # создаём пайплайы обработки
    cat_transformer = Pipeline(cat_steps)
    num_transformer = Pipeline(num_steps)
    # создаём общий пайплайн для подготовки данных
    data_preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features)], 
        remainder='passthrough')
    # создаём итоговый пайплайн по шагам
    step_prep = ('preprocessor', data_preprocessor)
    # выбор модели 
    if model_step == 'init':
        step_sol  = ('model', LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', penalty='l1', C=1))
    else: 
        step_sol  = ('model', DecisionTreeClassifier(random_state=RANDOM_STATE, max_depth=3, max_features=3, min_samples_leaf=4, min_samples_split=3))
    # собираем шаги модели
    pipe_final = Pipeline([step_prep, step_sol])
    
    return X_train, X_test, y_train, y_test, pipe_final
```

Сформируем входные параметры для модели. Так как была отмечена мультиколлинерность в значениях 'предыдущий_месяц_money' и 'текущий_месяц_money' - уберем "предыдущий", так как текущий - по-актуальнее.

```python
num_out_list = ['!Маркет_актив_тек_мес', '!Средний_просмотр_категорий_за_визит', '!Неоплаченные_продукты_штук_квартал', '!Ошибка_сервиса', 'предыдущий_месяц_money']
cat_out_list = ['Покупательская активность']
num_col= num_col_market_joined[~np.isin(num_col_market_joined, num_out_list)]
cat_col= cat_col_market_joined[~np.isin(cat_col_market_joined, cat_out_list)]
target = 'Покупательская активность'
```

В ручную переберем параметры кодирования и масштабирования (подбор параметров и модели используем в следующей пункте)

```python
## кодировщик: 'OneHot'/'Ordinal'
## маштабер:   'Standart'/'MinMax'
## настройки модели 'init'/'selected'
X_train, X_test, y_train, y_test, log_model = prediction_model(market_joined, num_col, cat_col, target, 'MinMax', 'Ordinal', 'init')
log_model.fit(X_train, y_train)
probability = log_model.predict_proba(X_test)[:,1]
y_test_predict = (probability>=0.5).astype(int)
print('Точность модели:',roc_auc_score(y_test_predict, y_test)) 
```

```python
def param_grid_choose(type_of_model):
# типы моделей DecisionTreeClassifier/KNeighborsClassifier/SVC/CatBoostClassifier/LogisticRegression
    RANDOM_STATE = 42
    if type_of_model == 'DecisionTreeClassifier':
        # словарь для модели DecisionTreeClassifier
        param_grid = [{
        'model': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'model__max_depth': range(2, 6),
        'model__min_samples_split': range(2, 6),
        'model__min_samples_leaf': range(1, 6),
        'model__max_features': range(2, 5),
#      , 'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
                     }]    
    elif type_of_model == 'KNeighborsClassifier':
        # словарь для модели KNeighborsClassifier 
         param_grid = [{
        'model': [KNeighborsClassifier()],
        'model__n_neighbors': range(2, 5)
#        , 'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
                      }]  
    elif type_of_model == 'SVC':
        # словарь для модели SVC
        param_grid = [{
        'model': [SVC(random_state=RANDOM_STATE)],
        'model__kernel':['rbf', 'linear'],
        'model__degree':[1,2],
        'model__C': [0.1,1,10,100],
        'model__gamma': [0.1,1,10,100]
#        , 'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
                    }]
    elif type_of_model == 'CatBoostClassifier':
        # словарь для модели CatBoostClassifier   
        param_grid = [{
        'model': [CatBoostClassifier(iterations=10,loss_function='CrossEntropy', logging_level ='Silent')],
        'model__learning_rate':[0.1,0.25, 0.5, 0.75 , 0.95],
#        , 'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
                    }]
    else:
        # словарь для модели LogisticRegression
        param_grid = [{
        'model': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', penalty='l1')],
        'model__C': [0.1,1,10,100]
#        , 'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
                   }]
    return param_grid
```

```python
X_train, X_test, y_train, y_test, test_log_model = prediction_model(market_joined, num_col,
                                                                    cat_col, target, 'MinMax', 'Ordinal', 'init')
```

```python
%%time
model_list = ['DecisionTreeClassifier', 'KNeighborsClassifier', 'SVC', 'CatBoostClassifier', 'LogisticRegression']
#model_list = ['DecisionTreeClassifier', 'KNeighborsClassifier']
metrics = []
for i, model in enumerate( model_list):
    randomized_search = RandomizedSearchCV(test_log_model, param_distributions = param_grid_choose(model), cv=5,
                                           scoring='roc_auc', random_state=RANDOM_STATE, n_jobs=-1);
    randomized_search.fit(X_train, y_train)
    y_test_pred = randomized_search.predict(X_test)
    train_score = randomized_search.best_score_
    test_score  = roc_auc_score(y_test, y_test_pred)
    shift  = round(1-train_score,2)*100
    spread =  round(abs(test_score - train_score),2)*100
    best_estimator = randomized_search.best_estimator_.get_params()["model"] 
#print('Лучшая модель и её параметры:', )
    metrics.append([train_score, test_score,  shift , spread, best_estimator])
metrics = pd.DataFrame(metrics)
metrics.columns = ['ROC_train ', 'ROC_test', 'Cмещения модели, %', 'Разброс модели, %', 'best_estimator']
metrics.index  = model_list
metrics
```

- Лучшие значения показала модель "CatBoostClassifier" ее не будем исследовать и применять, а использует как эталонную(чтобы было понятно куда стремится);
- Модель "SVC" выглядет переобученной на тренеровочной выборке;
- Лучше показатели по тестовой выборке (при обучении по 5 фолдам) показала DecisionTreeClassifier используем ее как итоговую.

С точки зрения выбранной модели, малый разброс говорит о приемлемой сложности модели, а уменьшить смещение возможно с использованием новых данных.
